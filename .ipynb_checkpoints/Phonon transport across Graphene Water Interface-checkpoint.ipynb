```lammps

group 		water type 1 2

group		Linterface_dump region Linterface_dump

group		Rinterface_dump_wat subtract Rinterface_dump gnc

#-------------------------------------------------------

#-------------------------------------------------------

#-------------------------------------------------------
#Production run
#-------------------------------------------------------
```

## Density of the water box

The average number density distribution of water molecules during simulation along the z-direction is calculated with the following code:

```tcl
package require pbctools

proc density {psfPrefix dcdPrefix timeStep dcdFreq } {
mol load psf $psfPrefix.psf
mol addfile $dcdPrefix.dcd waitfor all

set nFrame [molinfo top get numframes]
set timeGap [expr 1.0e-6*$timeStep*$dcdFreq];

pbc unwrap -all
set sel [atomselect top all]
set mm [measure minmax $sel]
$sel delete

puts $mm

set xlen [expr [lindex $mm 1 0] - [lindex $mm 0 0]]
set ylen [expr [lindex $mm 1 1] - [lindex $mm 0 1]]
set zlen [expr [lindex $mm 1 2] - [lindex $mm 0 2]]
puts $xlen
puts $ylen
puts $zlen

set xlen 35.0
set ylen 35.0

set nBin 100
set binWidthx [expr double($xlen/$nBin)]
set binWidthy [expr double($ylen/$nBin)]
set binWidthz [expr double($zlen/$nBin)]	

set binxVol [expr 0.001*$binWidthx*$ylen*$zlen]; # nm^3
set binyVol [expr 0.001*$binWidthy*$xlen*$zlen];
set binzVol [expr 0.001*$binWidthz*$xlen*$ylen];

for {set iBin 0} {$iBin < $nBin} {incr iBin} {
   set denX($iBin) {}
   set densx($iBin) 0 
   set denY($iBin) {}
   set densy($iBin) 0
   set denZ($iBin) {}
   set densz($iBin) 0
}

set sel [atomselect top "type OT"]
set n [$sel num]
$sel delete

for {set iFrame 1} {$iFrame < $nFrame} {incr iFrame} {
   
   molinfo top set frame $iFrame
   set sel [atomselect top "type OT"]
   set xvec [$sel get x]
   set yvec [$sel get y]
   set zvec [$sel get z]
   $sel delete

   for {set iBin 0} {$iBin < $nBin} {incr iBin} {
      set densx($iBin) {}
      set densx($iBin) 0 
      set densy($iBin) {}
      set densy($iBin) 0
      set densz($iBin) {}
      set densz($iBin) 0
   }

   for {set i 0} {$i < $n} {incr i} {
     set xp [lindex $xvec $i]
     set yp [lindex $yvec $i]
     set zp [lindex $zvec $i]

     set x [expr ($xp/$binWidthx)]
     set y [expr ($yp/$binWidthy)]
     set z [expr ($zp/$binWidthz)]
     set binxIndex [expr int($x)]
     set binyIndex [expr int($y)]
     set binzIndex [expr int($z)]

     incr densx($binxIndex)
     incr densy($binyIndex)
     incr densz($binzIndex)
   }
   for {set iBin 0} {$iBin < $nBin} {incr iBin} {
      lappend denX($iBin) [expr $densx($iBin)/$binxVol]
      lappend denY($iBin) [expr $densy($iBin)/$binyVol]
      lappend denZ($iBin) [expr $densz($iBin)/$binzVol]
   }
   
#   set fp [open ${iFrame}-Density.dat w]
#   for {set iBin 0} {$iBin < $nBin} {incr iBin} {
#      puts $fp [format "%.4f \t %.4f \t %.4f \t %.4f" $iBin $densx($iBin) $densy($iBin) $densz($iBin)]
#   }	
}

set fp [open avg-den.dat w]
for {set iBin 0} {$iBin < $nBin} {incr iBin} {
	set denAvgX [vecmean $denX($iBin)]
	set denErrX [expr [vecstddev $denY($iBin)]/sqrt(1000)]
	set denAvgY [vecmean $denY($iBin)]
	set denErrY [expr [vecstddev $denY($iBin)]/sqrt(1000)]
	set denAvgZ [vecmean $denZ($iBin)]
	set denErrZ [expr [vecstddev $denZ($iBin)]/sqrt(1000)]
	puts $fp [format "%.4f \t %.4f \t %.4f \t %.4f \t %.4f \t %.4f \t %.4f" $iBin $denAvgX $denAvgY $denAvgZ $denErrX $denErrY $denErrZ]

}

close $fp
mol delete all
}
```

## Post Processing


<blockquote>
The velocity data is obtained for group of graphene and water molecules seperately and are analysed with the following <b>bash<\b> script to seperate into different files for each atom.
</blockquote>



```bash
awk '/ITEM: ATOMS/,/ITEM: TIMESTEP/' velocity.dat > tmp.dat
sed -i '/ITEM/d' tmp.dat
mv tmp.dat vel.dat

sort -s -n -k1,1 vel.dat > velSorted.dat

awk 'NR%16667==0{print $1}' velSorted.dat > ids.dat

i=1
num=$(sed "${i}q;d" ids.dat)
while [ ${num} ]; 
do
	i=$(($i+1))
	awk '$1=='${num}'{print $0}' velSorted.dat > velxyz.tmp
	awk '{r=($2*$2+$3*$3+$4*$4)**0.5;}{printf"%d %0.20f %0.20f %0.20f %0.20f\n",$1,$2,$3,$4,r;}' velxyz.tmp > vel.tmp
	cp vel.tmp vel${num}.txt
	echo done $num 
	rm -f *.tmp	
	num=$(sed "${i}q;d" ids.dat)
done
```


<blockquote>

Once the velocity files are obtained from <b>lammps code #2</b> the following python script is used to calculate DOS from eqn:
</blockquote>


$$ DOS = fft \langle V \bigodot V \rangle $$

## Right Interface Graphene

````python
import numpy as np
import matplotlib.pyplot as plt
import scikits.talkbox.tools as tl
ids = np.loadtxt("ids_gra_R.dat")

tsteps = 16667
nvel = 6000
count = tsteps/nvel
n = 2**tl.nextpow2(nvel)
Dsum = 0
natoms = 50
Dt = 1
for j in range(count):
    Dx = 0
    Dy = 0
    Dz = 0
    for i in ids:
        val = int(i)
        fname = "velRgra"+str(val)+".txt"
        data = np.loadtxt(fname)
        Vx = data[nvel*(j):nvel*(j+1)-1,1]
        Vy = data[nvel*(j):nvel*(j+1)-1,2]
        Vz = data[nvel*(j):nvel*(j+1)-1,3]
        Dxsum = np.fft.fft(Vx,n)
        Dx = Dx + np.abs(Dxsum/n)**2
        Dysum = np.fft.fft(Vy,n)
        Dy = Dy + np.abs(Dysum/n)**2
        Dzsum = np.fft.fft(Vz,n)
        Dz = Dz + np.abs(Dzsum/n)**2
    print j   
    D = (Dx+Dy+Dz)/3/natoms
    Dsum = Dsum + D

DOS = Dsum/count
N = np.size(Dsum)
DOS_Gra_R = DOS[:N/2]
freq = np.fft.fftfreq(N,d = 0.012)
freq = freq
freq = freq[:N/2]
#freq = (1/(n*Dt*1e-15))*((0:np.size(DOS)-1)/1e15)
plt.xlabel("f(THz)")
plt.ylabel("DOS")
plt.plot(freq,DOS_Gra_R)
plt.show()
Dsum
````

### Combined DOS for Water and Graphene at Right Interface

````python
from pwtools import signal
import scipy

a = signal.smooth(DOS_Gra_R, scipy.signal.hann(50))
b = signal.smooth(DOS_Wat_R, scipy.signal.hann(50))
c = signal.smooth(DOS_Gra_L, scipy.signal.hann(50))
d = signal.smooth(DOS_Wat_L, scipy.signal.hann(50))
plt.plot(freq,a, label='Gra_R')
plt.plot(freq,b, label='Wat_R')
plt.plot(freq,c, label='Gra_L')
plt.plot(freq,d, label='Wat_L')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
````

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

vel = np.loadtxt('vel.tmp')
vel = vel[:,1]
corr = np.correlate(vel,vel,mode='full')
N = np.size(corr)
corr = corr[:N/2]
```
