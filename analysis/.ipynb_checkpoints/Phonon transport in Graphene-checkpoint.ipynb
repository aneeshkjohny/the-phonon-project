# Phonon transport in Graphene

### Objective 1

Reproducing the DOS graphs in Ref "[Phonon transmission and thermal conductance across graphene/Cu interface](http://aip.scitation.org/doi/10.1063/1.4821439)"


### Geometry script
```tcl
package require nanotube
package require topotools

set molname graphene
graphene -lx 10 -ly 10 -type zigzag -b 0
set sel [atomselect top all]
$sel moveby {0.7090 0.0 0.0}
$sel writepdb $molname.pdb
$sel writepsf $molname.psf
topo writelammpsdata $molname.lmpsys

set mm [measure minmax $sel]
set xlo [format %.4f [expr [lindex $mm 0 0]+0.0] ]
set xhi [format %.4f [expr [lindex $mm 1 0]+1.4200] ]; # very imp
set ylo [format %.4f [expr [lindex $mm 0 1]+0.0] ]
set yhi [format %.4f [expr [lindex $mm 1 1]+1.2298] ]; # very imp
set zlo [format %.4f [expr [lindex $mm 0 2]-1.0] ]
set zhi [format %.4f [expr [lindex $mm 1 2]+1.0] ]

exec sed -i "12s/.*/  $xlo $xhi  xlo xhi/" $molname.lmpsys
exec sed -i "13s/.*/  $ylo $yhi  ylo yhi/" $molname.lmpsys
exec sed -i "14s/.*/  $zlo $zhi  zlo zhi/" $molname.lmpsys

exec rm -f $molname.pdb
mol delete all
exit
```

## Post Processing


<blockquote>
The velocity file <b>vel.dat</b> is obtained and it is sorted and spereate velocity files are written for each atom using the following <b>bash</b> script:
</blockquote>



```bash
awk '/ITEM: ATOMS/,/ITEM: TIMESTEP/' velocity.dat > tmp.dat
sed -i '/ITEM/d' tmp.dat
mv tmp.dat vel.dat

sort -n -k1 vel.dat > velSorted.dat

awk 'NR%5000==0{print $1}' velSorted.dat > ids.dat     #Since data collected every 100 steps for 500000 steps
        
i=1
num=1
while [ ${num} ]; 
do
        awk '$1=='${num}'{print $0}' velSorted.dat > velxyz.tmp
        awk '{r=($2*$2+$3*$3+$4*$4)**0.5;}{printf"%d %0.20f %0.20f %0.20f %0.20f\n",$1,$2,$3,$4,r;}' velxyz.tmp > vel.tmp
        cp vel.tmp vel${num}.txt
        echo done $num 
        rm -f *.tmp
        num=$(sed "${i}q;d" ids.dat)
        i=$(($i+1))
done
```


<blockquote>

Once the velocity files are obtained from <b>lammps code #2</b> the following python script is used to calculate DOS from eqn:
</blockquote>


$$ DOS = fft \langle V \bigodot V \rangle $$

````python
import numpy as np
import matplotlib.pyplot as plt
import scikits.talkbox.tools as tl
ids = np.loadtxt("ids.dat")

tsteps = 10001
nvel = 1000
count = tsteps/nvel
n = 2**tl.nextpow2(nvel)
Dsum = 0
natoms = 3963
Dt = 1
for j in range(count):
    Dx = 0
    Dy = 0
    Dz = 0
    for i in ids:
        val = int(i)
        fname = "vel"+str(val)+".txt"
        data = np.loadtxt(fname)
        Vx = data[1+nvel*(j):nvel*(j+1),1]
        Vy = data[1+nvel*(j):nvel*(j+1),2]
        Vz = data[1+nvel*(j):nvel*(j+1),3]
        Dxsum = np.fft.fft(Vx,n)
        Dx = Dx + np.abs(Dxsum/n)
        Dysum = np.fft.fft(Vy,n)
        Dy = Dy + np.abs(Dysum/n)
        Dzsum = np.fft.fft(Vz,n)
        Dz = Dz + np.abs(Dzsum/n)
    print j   
    D = (Dx+Dy+Dz)/3/natoms
    Dsum = Dsum + D

DOS = Dsum/count
N = np.size(Dsum)
DOS = DOS[:N/2]
freq = np.fft.fftfreq(N,d = 0.001)
freq = freq
freq = freq[:N/2]
#freq = (1/(n*Dt*1e-15))*((0:np.size(DOS)-1)/1e15)
plt.xlabel("f(THz)")
plt.ylabel("DOS")
plt.plot(freq,DOS)
plt.show()
Dsum
````

The DOS calculations are now performed in the following way

* The velocity output files are used to calculate the velocity auto correlation(vacf).
* These values are then used to calculate DOS following the procedures in ref:
  "[Implications of cross-species interactions on the temperature dependence of Kapitza conductance](https://journals.aps.org/prb/abstract/10.1103/PhysRevB.84.193301)"
      

<blockquote>
The following code is used to calculate vacf
</blockquote>

```bash
i=1
num=1
while [ ${num} ]; 
do
        cp ./vel$num.txt vel.tmp
        python vacf.py
        cp vacf.tmp vacf${num}.dat
        echo "written vacf$num"
        rm -f *.tmp
        num=$(sed "${i}q;d" ids.dat)
        i=$(($i+1))
done

paste -d" " vacf{1..1000}.dat > vacf_combined1.tmp
paste -d" " vacf{1001..2000}.dat > vacf_combined2.tmp
paste -d" " vacf{2001..3000}.dat > vacf_combined3.tmp
paste -d" " vacf{3001..3936}.dat > vacf_combined4.tmp
awk '{tot=0;avg=0;}{for(i=1;i<=NF;i++){tot+=$i;}avg=tot/NF;printf "%.30f\n",avg;}' vacf_combined1.tmp > vacf_avg1.dat
awk '{tot=0;avg=0;}{for(i=1;i<=NF;i++){tot+=$i;}avg=tot/NF;printf "%.30f\n",avg;}' vacf_combined2.tmp > vacf_avg2.dat
awk '{tot=0;avg=0;}{for(i=1;i<=NF;i++){tot+=$i;}avg=tot/NF;printf "%.30f\n",avg;}' vacf_combined3.tmp > vacf_avg3.dat
awk '{tot=0;avg=0;}{for(i=1;i<=NF;i++){tot+=$i;}avg=tot/NF;printf "%.30f\n",avg;}' vacf_combined4.tmp > vacf_avg4.dat
paste -d" " vacf_avg* > vacf_avg.tmp
awk '{tot=0;avg=0;}{for(i=1;i<=NF;i++){tot+=$i;}avg=tot/NF;printf "%.30f\n",avg;}' vacf_avg.tmp > vacf_avg.dat
paste vacf_avg.dat > vacf.dat

```

<blockquote>
The above is done separately for x, y, and z directions with corresponding velocity arrays read in file "vacf.py" which is given below.
</blockquote>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

vel = np.loadtxt('vel.tmp')
vel = vel[:,1]
length = len(vel)
corr = np.correlate(vel,vel,mode='full')
N = np.size(corr)
corr = corr[N/2:]
for i in range(length):
        corr[i] = corr[i]/(length - i)
np.savetxt("vacf.tmp",corr)
```

<blockquote>
Now after obtaining vacf in x, y and z directions successfully, the DOS calculation is performed using the following code.
</blockquote>

### For 1000 point data (sampling time 10 ps)

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

velx = np.loadtxt('vacf10000x.dat')
vely = np.loadtxt('vacf10000y.dat')
velz = np.loadtxt('vacf10000z.dat')
corrx = np.correlate(velx,velx,mode='full')
corry = np.correlate(vely,vely,mode='full')
corrz = np.correlate(velz,velz,mode='full')
Nx = np.size(corrx)
Ny = np.size(corry)
Nz = np.size(corrz)
corrx = corrx[:Nx/2]
corry = corry[:Ny/2]
corrz = corrz[:Nz/2]
fx, Pxx_den = signal.welch(corrx,fs=1.0,window='hamming',nperseg=70,noverlap=35,nfft=120,detrend='constant', return_onesided=True)
fy, Pyy_den = signal.welch(corry,fs=1.0,window='hamming',nperseg=70,noverlap=35,nfft=120,detrend='constant', return_onesided=True)
fz, Pzz_den = signal.welch(corrz,fs=1.0,window='hamming',nperseg=70,noverlap=35,nfft=120,detrend='constant', return_onesided=True)
ax = [fx, Pxx_den]
ay = [fy, Pyy_den]
az = [fz, Pzz_den]
#plt.plot(fx,Pxx_den,label = 'Pxx')
#plt.plot(fy,Pyy_den,label = 'Pyy')
plt.plot(fz,Pzz_den,label = 'Pzz')
P = (Pxx_den + Pyy_den + Pzz_den)/3
plt.plot(fx,P,label='mean')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
#np.savetxt("test.txt",a)
plt.show()
````
