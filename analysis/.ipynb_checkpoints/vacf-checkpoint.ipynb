<h1>velocity autocorrelation
    $$ \langle V \bigodot V \rangle $$

```python
import numpy as np
import matplotlib.pyplot as plt

vel = np.loadtxt('vel1.txt')
vel = vel[:,1]
corr = np.correlate(vel,vel,mode='full')
N = np.size(corr)
corr = corr[:N/2]
freq = np.fft.rfftfreq(corr.shape[-1], d=0.02)
velFFT = np.fft.rfft(corr)
a = np.abs(velFFT)
plt.plot(freq,a)
plt.show()
np.size(a)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

vel = np.loadtxt('vel1.txt')
vel = vel[:,1]
corr = np.correlate(vel,vel,mode='full')
N = np.size(corr)
corr = corr[:N/2]
f, Pxx_den = signal.welch(corr,fs=1.0,window='hamming',nperseg=50,noverlap=25,nfft=200,detrend='constant', return_onesided=True)
a = [f, Pxx_den]
plt.plot(f,Pxx_den)
#np.savetxt("test.txt",a)
plt.show()
```

### The files vacf1 and vacf2 are for individual atoms.  
### __vacf.dat is combined__ 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

data = np.loadtxt("vacf.dat")
data1 = np.loadtxt("vacf1.dat")
data2 = np.loadtxt("vacf2.dat")
data3 = np.loadtxt("vacf3.dat")
data4 = np.loadtxt("vacf4.dat")
data5 = np.loadtxt("vacf5.dat")

plt.clf()

plt.plot(data, label ='vacf')
plt.plot(data1, label ='atom1')
plt.plot(data2, label ='atom2')
plt.plot(data3, label ='atom3')
plt.plot(data4, label ='atom4')
plt.plot(data5, label ='atom5')


plt.axis([0.0,10,-1000,2000])

plt.legend(bbox_to_anchor=(0.75, 0.99), loc=2, borderaxespad=0.)

plt.savefig("vacf.pdf")
```
